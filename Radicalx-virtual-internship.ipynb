```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import norm
style.use("ggplot")
%matplotlib inline
```

```python
data = pd.read_csv("./download/Internship_Data.csv")   
```

```python
data.head(10)
```

```python
data.info()
```

```python
cols = ["Country 1", "Name", "Squad Number"]
data = data.drop(cols, axis = "columns")
```

```python
data.head(10)
```

```python
new_data = data.copy()
```

```python
col0 = new_data.columns[0]
new_data.rename(columns = {col0 : "Language"}, inplace = True)
```

```python
new_data.head()
```

```python
print(new_data["Language"][200:400])
```

```python
popular = ["Chinese", "Mandarin", "English", "Spanish", "Arabic", "Bengali", "Hindi", "Russian", "Portuguese", "Japanese", "German", "Chinese", "Japanese", "French", "Turkish", "Telugu", "Chinese", "Tamil", "Italian", "Urdu", "Chinese", "Jinyu", "Gujarati", "Polish", "Ukrainian", "hakka", "Xiang", "Malayalam", "kannada", "oriya", "Sunda", "Panjabi", "Romanian", "Bhojpuri", "Maithili", "Burmese", "awadhi", "India", "Thai", "Yoruba", "Sindhi"]
```

```python
popular = [i.lower() for i in popular]
```

```python
langs = {}
for lang in popular:
    langs[lang] = 0
```

```python

for val in new_data["Language"]:
    #print("gv")
    d = str(val).split()
    for word in d:
        if word.lower() in popular:
            langs[word.lower()] += 1
    
```

```python
d = list(langs.keys())
for key in d:
    if langs[key] == 0:
        langs.pop(key)
```

```python
langs = sorted(langs.items(), key=lambda x: x[1], reverse = True) 
langs = dict(langs)
```

```python
print(list(langs.keys()))
print(list(langs.values()))
```

```python
plt.bar(list(langs.keys()), list(langs.values()))
plt.xticks(rotation = 90)
plt.show()
```

```python
col1 = new_data.columns[1]
print(col1)
new_data.rename(columns = {col1 : "Program"}, inplace = True)
```

```python
new_data["Program"].fillna("Empty", inplace = True)  
```

```python
programs = set()
for val in new_data["Program"]:
    d = str(val).split(",")
    for word in d:
        word = word.strip()
        programs.add(word.lower())
```

```python
programs
```

```python
prog = {}
for i in  programs:
    prog[i] = 0
for val in new_data["Program"]:
    d = str(val).split(",")
    for word in d:
        if word.lower() in programs:
            prog[word.lower()] += 1
```

```python
prog.values()
```

```python
prog = sorted(prog.items(), key=lambda x: x[1], reverse = True) 
prog = dict(prog)
prog.values()
```

```python
plt.bar(list(prog.keys()), list(prog.values()))
plt.xticks(rotation = 90)
plt.show()
```

```python
new_data["Age"].astype("int")
```

```python
# Plot pandas histogram from dataframe with df.plot.hist (not df.hist)
ax = new_data['Age'].plot.hist(bins=20, density=True, edgecolor='w', linewidth=0.5)

# Save default x-axis limits for final formatting because the pandas kde
# plot uses much wider limits which usually decreases readability
xlim = ax.get_xlim()

# Plot pandas KDE
new_data['Age'].plot.density(color='k', alpha=0.5, ax=ax) # same as df['var'].plot.kde()

# Reset x-axis limits and edit legend and add title
ax.set_xlim(xlim)
ax.set_ylabel("Density")
ax.set_title('Age distribution of applicants', fontsize=14, pad=15)

plt.show()
```
